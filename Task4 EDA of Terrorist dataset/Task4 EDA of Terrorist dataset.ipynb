# Handling Missing Values in dataset

```python
def missing_values_table(df):
        # Total missing values
        mis_val = df.isnull().sum()
        mis_val_percent = 100 * df.isnull().sum() / len(df)        
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        
        return mis_val_table_ren_columns
```

```python
pd.set_option("max_rows", None)

#get the each feature with percentage of missing values from dataset
missing_value_df = missing_values_table(df)
missing_value_df
```

```python
#drop the features which has more than 50% missing values
delete_columns = missing_value_df[missing_value_df.iloc[:,1]>50].index

df_cleaned_missingvalue_column = df.drop(delete_columns,axis=1) 
```

```python
#Renaming the columns
df_cleaned_missingvalue_column.rename(columns={'iyear':'Year','imonth':'Month','extended':'Extended','iday':'Day','country_txt':'Country',
                   'provstate':'State','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target',
                   'nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                   'weaptype1_txt':'Weapon_type','weapdetail':'Weapon_detail','motive':'Motive','city':'City'
                               ,'latitude':'Latitude','longitude':'Longitude','natlty1_txt':'nationality'},inplace=True)
df_cleaned_missingvalue_column.head()
```

# Analyze the dataset by visualization graphs
